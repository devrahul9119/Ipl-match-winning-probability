```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('IPL Matches 2008-2020.csv')
delivery = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
total_runs_df = delivery.groupby(['id','inning'])['total_runs'].sum().reset_index()
```

```python
total_runs_df['total_runs'] = total_runs_df['total_runs']+1
```

```python
total_runs_df = total_runs_df[total_runs_df['inning'] == 1]
```

```python
match_df = match.merge(total_runs_df[['id','total_runs']],left_on = 'id',right_on = 'id')
```

```python
match_df['team1'].unique()
```

Teams with maximum matches

```python
teams = ['Royal Challengers Bangalore',
         'Kings XI Punjab', 
         'Mumbai Indians', 
         'Kolkata Knight Riders',
         'Rajasthan Royals', 
         'Chennai Super Kings',
         'Sunrisers Hyderabad',
          'Delhi Capitals']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

Filerting the matches which are not affected by rain or any other reasons

```python
match_df = match_df[match_df['neutral_venue']==0]
```

```python
match_df = match_df[['id','city','total_runs','winner']]
```

```python
delivery_df = match_df.merge(delivery,on= 'id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning']==2]
```

```python
delivery_df['current_score'] = delivery_df.groupby('id')['total_runs_y'].cumsum()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x== '0' else '1').astype(int)
```

```python
wickets = delivery_df.groupby('id')['player_dismissed'].cumsum().values
```

```python
delivery_df['wickets'] = 10 - wickets
```

```python
# current Runrate : runs/overs
delivery_df['current_run_rate'] = delivery_df['current_score']*6/(120 - delivery_df['balls_left'])
```

```python
# required Runrate = runsleft/overs_left
delivery_df['req_run_rate'] = delivery_df['runs_left']*6/delivery_df['balls_left']
```

```python
def result(data):
    return 1 if data['batting_team'] == data['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis = 1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','current_run_rate','req_run_rate','result','total_runs_x']]
```

```python
# Shuffling the dataset
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final_df.drop(columns = ['result']),final_df['result'],test_size = 0.2,random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
transformer = ColumnTransformer([('trf',OneHotEncoder(sparse = False,drop = 'first'),['batting_team','bowling_team','city'])],remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps = [('step1',transformer),('step2',LogisticRegression(solver = 'liblinear'))])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[1]
```
